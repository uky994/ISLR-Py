If we want to generate a vector of random variables from a normal distribution with mean  $\mu$  and standard deviation  $\sigma$ , we can use the function `np.random.normal(loc, scale, size)` where size is the sample size,loc is the mean $\mu$  which defaults to 0 and, and scale is the standard deviation  $\sigma$  which defaults to 1. If no value is given for size then just a single number is returned. 

The `numpy.random.normal` function generates an array or scalar of random normal variables, with first argument `loc` the mean of the distribution.  Each time we call this function, we will get a different answer. Here we create two correlated sets of numbers, `x` and `y`, and use the `corrcoef()` function to compute the correlation between them.

```python
x = np.random.normal(0, 1, 50)
y = x + np.random.normal(loc=50, scale=.1, size=50)

np.corrcoef(x, y)
```

By default, `numpy.random.normal` creates standard normal random variables with a mean of 0 and a standard deviation of 1. However, the mean and standard deviation can be altered using the `loc` and `scale` arguments, as illustrated above.
Sometimes we want our code to reproduce the exact same set of random numbers; we can use the `np.random.seed()` function to do this. The `np.random.seed()` function takes an (arbitrary) integer argument. Should we want to reproduce the exact same set of random numbers each time our code is run, we can use the function `np.random.seed(n)`, where *n* is an integer, to set the seed for our random number generator.

```python
np.random.seed(1303)
np.random.normal(size=5)
```

We use `np.random.seed()` throughout the labs whenever we perform calculations involving random quantities. In general this should allow the user to reproduce our results. However, small
discrepancies may arise depending on our versions of `Python`.

The `mean()` and `var()` functions can be used to compute the mean and variance of an array of numbers. Applying `sqrt()` to the output of `var()`
will give the standard deviation. Or we can simply use the `std()` function.

```python
np.random.seed(3)
y=np.random.normal(size=100)
np.mean(y)
```

```python
np.var(y)
```

```python
np.power(np.var(y),0.5)
```

```python
np.std(y)
```

We will now create some more sophisticated plots. The `contour()` function produces a _contour plot_ in order to represent three-dimensional data; it is like a topographical map. It takes three arguments:

1. A vector of the `x` values (the first dimension),
2. A vector of the `y` values (the second dimension), and
3. A matrix whose elements correspond to the `z` value (the third dimension) for each pair of (x, y) coordinates.

As with the `plot()` function, there are many other inputs that can be used to fine-tune the output of the `contour()` function. 

```python
x = np.linspace(-np.pi, np.pi)
y = np.linspace(-np.pi, np.pi)
def f(x, y):
    return np.cos(y)/(1 + x**2)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
plt.contour(X, Y, Z)
```

We can also specify that we want more levels to be drawn by using the levels optional parameter.

```python
plt.contour(X, Y, Z, levels = 45)
```

We can also turn our contour plot into a heatmap by switching it to a filled contour plot via the `plt.contourf()` function. In addition, we'll ad a `plt.colorbar()` command to automatically generate an additional axis with labeled color information for the plot.

```python
plt.contourf(X, Y, Z, levels = 45)
plt.colorbar()
```

In order to do 3D plotting using Matplotlib, we need to import the `mplot3d` toolkit.

Now let's make a three-dimensional contour plot using our previously-defined  $X$ ,  $Y$ , and  $Z$ .

We can also change the viewing angle using the `view_init` method to set the elevation angle (angle relative to the *x*-*y* plane) and azimuthal angle (angle rotated counter-clockwise about the *z*-axis).

### Slicing NumPy Arrays

In this section, we discuss how to slice NumPy arrays to access subsets of the data stored within them. For example, let's suppose our data is stored in the two-dimensional NumPy array A.

```python
A = np.arange(16).reshape(4, 4)
print(A)
```

To access a single element, we can use a comma-separated tuple of indices `A[i, j]` where *i* is the row index and *j* is the column index. Note that NumPy arrays are zero-indexed and that we can also use negative indices to index from the end of the array.

```python
A[0, 0]
```

```python
A[1, 2]
```

```python
A[2, -1]
```

If we wish to slice an array, we can use the usual slicing syntax for standard Python lists: `A[row_start:row_stop:row_step, col_start:col_stop:col_step]`. Note that array slices are views rather than *copies* of the array data. This means that we modifying a slice will modify those entries in the original array.

```python
A[0:2, 1:3] #Access the first two rows and the 2nd+3rd columns
```

```python
A[0:2, :] #Access the first two rows
```

```python
A[:,0:2] #Access the first two columns.
```

As mentioned before, A.shape returns the dimensions of the array as a tuple (rows, columns).

```python
A.shape
```

Also note that we can reshape an array (as long as the product of the dimentions remains same) using the command `A.reshape()`.

```python
B = A.reshape((2, 8))
print(B)
```

```python
sns.displot(Auto["mpg"], kde = False, bins = 20)
```

If we want to produce a scatterplot matrix, i.e. a scatterplot of every pair of variables of a given data set, we can use a `PairGrid`. To do so, we first initialize the grid with our desired data set, and then pass the plotting function to a map method that will be called on each subplot. Alternatively, we can use the companion function, `pairplot()` that trades off some flexibility for faster plotting. The `pairplot()` function creates a scatterplot matrix, i.e. a scatterplot for every pair of variables. We can also produce scatterplots for just a subset of the variables.

```python
g = sns.PairGrid(Auto)
g.map(plt.scatter)
```

```python
sns.pairplot(Auto[['mpg', 'displacement', 'horsepower', 'weight', 'acceleration']])
```

Note that by default the `pairplot()` function produces histograms on the diagonal instead of scatterplots. If we wish to reproduce this behavior using a `PairGrid`, we need to use the `map_diag()` and `map_offdiag()` functions. Do note that the axis ticks won't correspond to the count or density axis of the plots on the diagonal.

```python
g = sns.PairGrid(Auto, vars = ["mpg", "displacement", "horsepower", "weight", "acceleration"])
g.map_offdiag(plt.scatter)
g.map_diag(plt.hist)
```
